```python
import pickle as pkl
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
with open('../tuning/54kev/seq_length.pkl', 'rb') as f:
    data = pkl.load(f)
```

```python
for dat in data:
    print(dat['value'], len(dat['metrics']))
```

```python
for dat in data:
    seq_length = dat['value']
    metrics = dat['metrics']
    losses = [x['val_loss'] for x in metrics]
    rscores = [x['rscore'] for x in metrics] 
    
    plt.plot(range(len(rscores)), rscores, label=f"{seq_length} sequence length")
plt.title("rscore vs epochs")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np


# -------------------------
mpl.rcParams['font.size'] = 13
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.titleweight'] = 'normal'

# -------------------------
#  Figure
# -------------------------
plt.figure(figsize=(9, 5))

# Consistent color cycle
colors = plt.cm.tab10(np.linspace(0, 1, 10))

# -------------------------
#  Plot all sequence lengths
# -------------------------
for idx, dat in enumerate(data):
    seq_length = dat['value']
    metrics = dat['metrics']

    rscores = np.array([x['rscore'] for x in metrics])

    plt.plot(
        rscores,
        label=f"{seq_length}",
        linewidth=2.0,
        color=colors[idx % 10],
        marker='o',
        markersize=5,
        markeredgecolor='black'
    )

# -------------------------
#  Labels ONLY (no title)
# -------------------------
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Validation R", fontsize=14)

# -------------------------
#  Grid (clean academic)
# -------------------------
plt.grid(which='major', linestyle='--', linewidth=0.6, alpha=0.6)

# -------------------------
#  Legend
# -------------------------
plt.legend(
    title="Seq Length",
    fontsize=12,
    title_fontsize=12,
    loc="center left",
    bbox_to_anchor=(1.02, 0.5)
)

plt.tight_layout()
plt.savefig("seq_length_tune_ch3.png", dpi=300, bbox_inches='tight')
plt.show()
```
