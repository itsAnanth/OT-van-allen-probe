```python
import os
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
with open('OT-van-allen-probe/tuning/seq_length.pkl', 'rb') as f:
    data = pkl.load(f)
```

```python
data
```

```python
for dat in data:
    seq_length = dat['value']
    metrics = dat['metrics']
    losses = [x['loss'] for x in metrics]
    rscores = [x['rscore'] for x in metrics] 
    
    plt.plot(range(len(rscores)), rscores, label=f"{seq_length} sequence length")
plt.title("Acurracy vs epochs")
plt.legend()
plt.show()
```
