```python
import torch


rscores = []
for i in range(1, 20):
    print(f"Loading checkpoint no. {i}")
    data = torch.load(f'../checkpoints/54kev_{i}.pth', map_location=torch.device('cpu'), weights_only=True)
    rscores.append(data['metrics']['rscore'])
```

```python
rscores
```

```python
import matplotlib.pyplot as plt

plt.plot(range(len(rscores)), rscores)
plt.xticks(range(0, len(rscores)))
plt.grid()
plt.title('rscores vs epochs')
plt.show()
```
